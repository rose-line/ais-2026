## Python - Dictionnaires

---

- Les indices d'une **liste** sont des entiers, tandis que les indices d'un **dictionnaire** peuvent être de n'importe quel type de données immuable (ex : chaînes de caractères, tuples, etc.)

  - on appelle ces indices des **clés** (keys), et une clé associée à sa valeur forment une **paire clé-valeur** (_key-value pair_)

- Comme une liste, un dictionnaire est **mutable**

- Comme une liste, un dictionnaire est une **structure de données**, mais **non ordonnée**

  - elle est spécialisée dans le stockage de données **associatives**, c'est-à-dire des données naturellement organisées en paires clé-valeur

---

### Le type `dict`

- Un dictionnaire est défini par une série de paires clé-valeur, séparées par des virgules, et entourées de accolades `{}`

  - ex. : `personne = {"nom": "Alice", "âge": 30, "ville": "Paris"}`

  - les clés doivent être uniques ; elles sont ici : `"nom"`, `"âge"` et `"ville"`

  - les valeurs peuvent être dupliquées ; elles sont ici : `"Alice"`, `30` et `"Paris"` (respectivement) ; les types des valeurs peuvent donc être différents (comme dans une liste, mais c'est beaucoup plus courant avec les dictionnaires)

- On accède à une valeur non plus par un indice entier (comme pour les listes), mais par sa clé, toujours avec la syntaxe utilisant des crochets :

  - ex. : `personne["nom"]` renvoie `"Alice"`

---

### Dictionnaire vs Liste

- Les dictionnaires sont non ordonnés, contrairement aux listes

  - l'ordre d'insertion des éléments n'a donc pas d'importance pour les dictionnaires

  - cela signifie que l'ordre dans lequel les éléments sont ajoutés à un dictionnaire n'est pas nécessairement l'ordre dans lequel ils seront stockés ou récupérés, et donc traités

  - les éléments ne peuvent donc pas être « slicés », et leur parcours ne garantit pas un ordre particulier, même à travers deux exécutions différentes du même programme

  - Accès à une clé qui n'existe pas ? => `KeyError` (similaire à `IndexError` pour les listes)

```python
annivs = {'Alice': '01/04', 'Bob': '12/12', 'Eve': '30/04'}
while True:
  print('Entrez un nom (\'Return\' pour quitter) : ')
  nom = input()
  if nom == '':
    break

  if nom in annivs:
    print(annivs[nom] + ' est l\'anniv de ' + nom)
  else:
    print('Je n\'ai pas l\'info pour ' + nom)
    print('Quel est son anniversaire ?')
    jour = input()
    annivs[nom] = jour
    print('Base de données des anniversaires mise à jour.')
```

### Liste de clés et de valeurs

- Trois méthodes permettent d'obtenir les clés, les valeurs ou les paires clé-valeur d'un dictionnaire :

  - `dict.keys()` : renvoie une vue des clés du dictionnaire

  - `dict.values()` : renvoie une vue des valeurs du dictionnaire

  - `dict.items()` : renvoie une vue des paires clé-valeur du dictionnaire, sous forme de tuples `(clé, valeur)`

- Ces "vues" ne sont pas des vraies listes : elles ne peuvent pas être modifiées directement, mais elles reflètent les changements apportés au dictionnaire d'origine

  - ex. : si on ajoute une nouvelle paire clé-valeur à `personne`, les vues obtenues précédemment avec `personne.keys()`, `personne.values()` ou `personne.items()` seront automatiquement mises à jour pour inclure cette nouvelle paire

- On peut parcourir directement ces vues avec une boucle `for` pour accéder à leurs éléments

```python
dico = {'couleur': 'rouge', 'age': 42}

print('\nClés :')
for cle in dico.keys() :
  print(cle)

print('\nValeurs :')
for val in dico.values() :
  print(val)

print('\nPaires (clé, valeur) :')
for cle, val in dico.items() :
  print(cle, val)

vue_valeurs = dico.values()
dico['age'] = 43
print('\nLa vue définie auparavant "voit" la nouvelle valeur :')
for val in vue_valeurs :
  print(val)

print('Mais je ne peux pas modifier une valeur à partir de la vue :')
vue_valeurs[0] = 'bleu'  # Provoque une erreur
```

- Pour obtenir une "vraie" liste à partir de ces vues, on peut les convertir avec la fonction `list()` :

  - ex. : `list(personne.keys())` renvoie une liste des clés du dictionnaire `personne`

- On peut utiliser les opérateurs `in` et `not in` pour vérifier la présence d'une clé ou d'une valeur dans un dictionnaire

  - `"nom" in personne` renvoie `True`, tandis que `"âge" not in personne` renvoie `False`

  - `"Alice" in personne.values()` renvoie `True`, et `40 not in personne.values()` renvoie `True`

### Vérifier la présence d'une clé

- Parfois on n'est pas sûr qu'une clé existe effectivement dans un dictionnaire ; on peut donc utiliser `in` ou `not in` pour tester la présence d'une clé avant d'y accéder, pour éviter un `KeyError` :

  - `if "nom" in personne: print(personne["nom"])`

- Ou alors utiliser une valeur par défaut si jamais la clé n'existe pas :

  - `if "salaire" in personne: print(personne["salaire"]) else: print(0)`

- Mais Python fournit une méthode spéciale pour ce type d'accès : la méthode `get()`, qui prend deux arguments :

  - la clé à rechercher

  - une valeur par défaut à retourner si la clé n'existe pas (optionnelle, `None` par défaut)

  - ex. : `personne.get("salaire", 0)` renvoie `0` si la clé `"salaire"` n'existe pas dans le dictionnaire `personne`, sinon elle renvoie la valeur associée à cette clé

  - on peut aussi utiliser `get()` sans second argument, auquel cas elle renverra `None` si la clé n'existe pas

```python
inventaire = {'pommes': 10, 'oranges': 5, 'bananes': 8}
print('Nombre de pommes dans l\'inventaire :', inventaire.get('pommes', 0))
print('Nombre de poires dans l\'inventaire :', inventaire.get('poires', 0))
```

### Ajout uniquement si la clé n'existe pas déjà

- Parfois on veut ajouter une paire clé-valeur à un dictionnaire, mais seulement si la clé n'existe pas déjà (pour éviter d'écraser une valeur déjà renseignée)

```python
personne = {'nom': 'Alice', 'âge': 30}
print(personne)
if ('ville' not in personne) :
  personne['ville'] = 'Paris'
print(personne)
if ('nom' not in personne) :
  personne['nom'] = 'Bob'
print(personne)
```

- La méthode `setdefault()` permet de faire cela en une seule opération :

  - `dict.setdefault(clé, valeur)` : si la clé existe déjà dans le dictionnaire, elle renvoie sa valeur déjà associée ; sinon, elle ajoute la paire clé-valeur au dictionnaire et renvoie la nouvelle valeur (appelée la « valeur par défaut »)

  - ex. : `personne.setdefault("ville", "Paris")` ajoute la paire `"ville": "Paris"` au dictionnaire `personne` seulement si la clé `"ville"` n'existe pas déjà, et renvoie la valeur associée à la clé `"ville"` (soit la valeur existante, soit `"Paris"` si elle a été ajoutée)

```python
personne = {'nom': 'Alice', 'âge': 30}
print(personne)
personne.setdefault('ville', 'Paris')
print(personne)
personne.setdefault('nom', 'Bob')
print(personne)
```

- L'exemple suivant utilise `setdefault()` pour un script de comptage de lettres dans un texte ; il parcourt chaque mot du texte, et utilise `setdefault()` pour, si nécessaire, initialiser le compteur de chaque lettre à `0` avant de l'incrémenter :

```python
message = 'Les dictionnaires sont très utiles pour compter des éléments dans un texte.'
frequence_lettres = {}  ## Dictionnaire vide, stockera les paires (lettre, occurrences)
for lettre in message:
    frequence_lettres.setdefault(lettre, 0)
    frequence_lettres[lettre] = frequence_lettres[lettre] + 1
print(frequence_lettres)
```

### Dictionnaires et listes imbriqués

- Lorsque les données sont plus complexes, on peut imbriquer des dictionnaires et des listes les uns dans les autres pour créer des structures de données plus riches et plus flexibles

  - les listes sont pratiques pour stocker des séries de valeurs ordonnées

  - les dictionnaires pour tout ce qui se modélise naturellement en paires clé-valeur

- Le programme suivant utilise un dictionnaire qui contient lui-même des dictionnaires de vivres que des invités apportent

  - la fonction `total_vivres()` calcule le total de chaque type de vivres apporté par tous les invités, en parcourant les dictionnaires imbriqués et en additionnant les quantités correspondantes


```python
apports_invites = {'Alice': {'pommes': 5, 'sandwiches jambon': 12},
                  'Bob': {'sandwiches jambon': 8, 'poires': 2, 'dessert': 5},
                  'Eve': {'dessert': 9, 'poires': 12}}

def inventaire_vivres(apports):
    inventaire = {}
    for apport_invite in apports.values():
        for aliment, nb in apport_invite.items():
            inventaire[aliment] = inventaire.get(aliment, 0) + nb
    return inventaire

print('Apports de chaque invité : ', apports_invites)
print('Inventaire total des vivres : ', inventaire_vivres(apports_invites))
```

### Synthèse

- Les listes et les dictionnaires sont tous deux des structures de données fondamentales, chacune avec ses propres caractéristiques et utilisations

- Les dictionnaires sont particulièrement utiles pour stocker des données **associatives** (paires clé-valeur), sans ordre naturel

- Les clés des dictionnaires doivent être uniques et immuables, partant de là elles peuvent être de n'importe quel type de données immuable (strings, tuples, nombres...)

- On peut imbriquer des listes dans des dictionnaires, et des dictionnaires dans des listes, et cela si nécessaire sur plusieurs niveaux pour créer des structures de données complexes et adaptées à des besoins spécifiques
