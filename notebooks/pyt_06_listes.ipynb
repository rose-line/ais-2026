## Python - Listes et Tuples

---

- Les **listes** et les **tuples** permettent de stocker des **collections de données ordonnées**

- Ce sont des **structures de données** essentielles pour pouvoir manipuler de grands ensembles d'éléments

- Comme les listes peuvent elles-mêmes contenir d'autres listes, elles permettent de représenter des structures de données hiérarchiques complexes (ex : tableaux à plusieurs dimensions)

---

### Le type `list`

- Une **liste** contient des _valeurs multiples ordonnées_

  - une liste est une _valeur_ elle-même : elle peut être stockée dans une variable comme toute autre valeur ; c'est juste que _cette valeur contient une collection d'autres valeurs_

  - les valeurs stockées dans une liste sont appelées des **éléments** (on dit aussi ***items***)

- On utilise des crochets `[ ]` pour définir une liste, et les éléments sont séparés par des virgules :

````python
# Une liste contient souvent des éléments du même type (entiers, flottants, strings...)
list1 = [34, 12, 56]
list2 = ['pomme', 'banane', 'cerise']

# Mais ce n'est pas obligatoire
list3 = ['pomme', 12.56, False, None, 10]

# Une liste peut être vide
list4 = []

# Un print sur une liste affiche bien la liste, avec la même notation
print(list2)         # ['pomme', 'banane', 'cerise']
````

---

### Les indices de liste (_index_)

- On a vu qu'une liste était une collection ordonnée : chaque élément a une position précise dans la liste

  - on dit que les éléments sont **indexés**

  - on appelle la position d'un élément son **indice** (ou **index**)

  - les listes sont indexées **à partir de 0** : le premier élément a l'indice 0, le deuxième l'indice 1, le dernier l'indice `n-1` si la liste contient `n` éléments

- On accède à un élément d'une liste en utilisant son indice entre crochets après le nom de la liste : `ma_liste[indice]`

````python
liste = ['saucisses', 'ketchup', 'frites']
print(liste[0])
print(liste[1])
combo = liste[2] + ' avec du ' + liste[1]
print(combo)

# On a une erreur 'IndexError' si on accède à un index trop grand
print(liste[4])        # 'IndexError: liste index out of range'
````

- Si une liste contient elle-même d'autres listes (tout à fait possible puisqu'une liste est une valeur comme une autre), on accède à une liste imbriquée en utilisant son indice comme d'habitude, et à un de ses éléments en ajoutant un second indice : `ma_liste[indice][indice_element]`

````python
liste = ['toto', ['pomme', 'banane', 'cerise'], [5, 4, 3]]
print(liste[0])    # 'toto'
print(liste[1])    # ['pomme', 'banane', 'cerise']
print(liste[1][0]) # 'pomme' - on accède à liste de fruits par liste[1], puis à son élément 0
print(liste[2][1]) # 4 - élément en position 1 de l'élément en position 2
````

#### Indices négatifs

En Python, vous verrez aussi parfois l'utilisation d'indices négatifs pour accéder aux éléments d'une liste en partant de la fin :

````python
liste = [0, 10, 20, 30, 40, 50]
print(liste[-1])            # dernier élément, équivalent ici à liste[5]
print(liste[-3])
````

#### Slices

- Un _slice_ (tranche) permet d'extraire directement une sous-liste d'une liste en spécifiant un intervalle d'indices

- La syntaxe utilise deux points `:` pour séparer l'indice de début et l'indice de fin : `ma_liste[indice_debut:indice_fin]`

  - c'est un intervalle « fermé - ouvert » : l'élément à l'indice de début est inclus, mais pas celui à l'indice de fin

  - conséquence : pour avoir rapidement la taille du slice, il suffit de faire `indice_fin - indice_debut` (sauf si indices négatifs, ça ne fonctionne pas)

````python
liste = ['Bruce', 'Steve', 'Dave', 'Adrian', 'Nicko']
print(liste[0:5])   # on commence à l'indice 0 et finit à l'indice 4
print(liste[1:3])   # 3-1=2 => la tranche commence à 1 et a une taille de 2
print(liste[0:-1])  # début jusqu'à juste avant le dernier élément (-2)
````

- On peut omettre l'indice de début ou de fin :

  - si on omet l'indice de début, le slice commence au début de la liste

  - si on omet l'indice de fin, le slice va jusqu'à la fin de la liste

  - si on omet les deux, le slice contient tous les éléments de la liste

````python
liste = ['Bruce', 'Steve', 'Dave', 'Adrian', 'Nicko']
print(liste[:3])    # début jusqu'à index 2
print(liste[3:])    # à partir de index 3 jusqu'à la fin
print(liste[:])     # toute la liste
````

#### Copie de liste

- Un slice crée donc une **nouvelle liste** qui est une copie des éléments extraits de la liste d'origine

- Il est parfois pratique de créer une **copie d'une liste** sans modifier l'originale

  - une technique simple est pratique est donc d'utiliser un slice qui couvre toute la liste : `copie = ma_liste[:]`

````python
liste = ['Bruce', 'Steve', 'Dave', 'Adrian', 'Nicko']
list_copy = liste[:]     # copie complète de la liste
list_copy[4] = 'Simon'  # modification de la copie
print(list_copy)        # copie modifiée
print(liste)             # liste originale est inchangée
````

### Mise à jour de liste

- Une opération très courante est de modifier un élément d'une liste

- Exactement comme on utilise l'affectation ('=') pour modifier la valeur d'une variable, on modifie ainsi un élément d'une liste : `ma_liste[indice] = nouvelle_valeur`

- Il ne s'agit pas ici d'une copie : c'est bien la liste originale qui est modifiée

````python
liste = ['Bruce', 'Steve', 'Dave', 'Adrian', 'Nicko']
liste[-1] = 'Simon'
print(liste)
````

- Notez que la _valeur de la liste_ elle-même (`liste`) _ne change pas_ : elle désigne toujours la même liste, c'est juste que les _éléments de la liste se mettent à jour_

### Autres opérations sur les listes

#### Concaténation

- On peut concaténer (joindre) deux listes (comme pour des strings) avec l'opérateur `+` : `liste3 = liste1 + liste2`

````python
liste1 = [1, 2, 3]
liste2 = [4, 5]
liste3 = liste1 + liste2
print(liste3)
````

#### Réplication

- Une opération moins courante consiste à répéter une liste plusieurs fois avec l'opérateur `*` : `liste2 = liste1 * n` crée une **nouvelle liste** qui contient les éléments de `liste1` répétés `n` fois de suite

````python
liste1 = [1, 2, 3]
liste2 = liste1 * 3
print(liste2)
````

#### Supprimer des éléments

- L'instruction `del` permet de supprimer un élément d'une liste en spécifiant son indice : `del ma_liste[indice]`

  - les éléments suivants sont décalés d'une position vers l'avant pour combler le trou laissé par l'élément supprimé

````python
liste1 = [1, 2, 3, 4, 5]
del liste1[1]      # on supprime le deuxième élément
print(liste1)      # reste 4 éléments
````

### Les listes : pour quoi faire ?

- Pour gérer de multiples valeurs qui représentent des concepts similaires (par exemple, des noms de fichiers, des adresses IP, des températures mesurées, etc.), on pourrait être tenté d'utiliser plusieurs variables :

```python
ip1 = "192.168.1.2"
ip2 = "192.168.10.3"
ip3 = "192.168.10.8"
```

- Cependant, cela peut rapidement devenir ingérable, surtout quand on ne connaît pas à l'avance le nombre d'éléments à traiter (cas fréquent)

- Les listes offrent une solution plus élégante et flexible, en permettant de regrouper au sein d'une seule variable une collection d'éléments :

```python
ips = ["192.168.1.2", "192.168.10.3", "192.168.10.8"]
```

- Si le nombre d'éléments à gérer change, on n'a pas besoin de modifier le code qui les traite : on adapte simplement la liste

- De plus, on peut facilement parcourir les éléments d'une liste avec une boucle pour effectuer des opérations sur chaque élément

#### Exemple

- Reprenons l'exemple des adresses IP : imaginons que l'on souhaite stocker des adresses IP récoltées sur un réseau

  - Si on le fait via des variables individuelles, on doit écrire du code spécifique pour chaque variable :

````python
print('Entrez l\'adresse IP 1 :')
ip1 = input()
print('Entrez l\'adresse IP 2 :')
ip2 = input()
print('Entrez l\'adresse IP 3 :')
ip3 = input()
print('Entrez l\'adresse IP 4 :')
ip4 = input()
print('Les adresses IP sont:')
print(ip1, ip2, ip3, ip4, sep=' ; ')
````

- Cela pose deux problèmes :

  - le code est répétitif et peu élégant

  - si le nombre d'adresses IP change, il faut modifier le code

- En utilisant une liste, on peut écrire un code plus concis et flexible :

````python
ips = []
while True:
  nb_ips = len(ips) + 1
  print('Entrez l\'adresse IP ' + str(nb_ips) + ' (appuyez sur Entrée pour arrêter) :')
  ip = input()
  if ip == '':
    break
  ips = ips + [ip]      # Concaténation de listes
print('Les adresses IP sont:')
print(ips, sep=' ; ')
````

- La liste est ce qu'on appelle une **structure de données** : un moyen d'organiser et de stocker des données de manière efficace pour pouvoir les manipuler facilement

### Parcours de liste

- La boucle `for` est idéale pour parcourir les éléments d'une liste

  - d'ailleurs, quand on écrit `for i in range(5)`, on parcourt en fait une liste virtuelle des nombres de 0 à 4 générée par `range(5)` : `[0, 1, 2, 3, 4]`

#### Parcours avec `range(len(...))`

- Pour parcourir explicitement une liste, un première option est d'utiliser `range(len(ma_liste))` :

  - `len(ma_liste)` calcule la longueur de la liste

  - on rappelle qu'une liste est indexée à partir de 0, donc les indices vont de 0 à `len(ma_liste) - 1`
  
  - `range` est donc naturellement utilisé pour générer les indices

````python
ips = ['192.168.1.10', '192.168.1.11', '192.168.1.12']
for i in range(len(ips)):     # génère les indices [0, 1, 2]
  print('Indice ' + str(i) + ' : ' + ips[i])
````

#### Parcours direct des éléments avec `in`

- Si on n'a pas besoin des indices, on peut parcourir les éléments de la liste en utilisant l'opérateur `in` **directement** sur la liste :

````python
ips = ['192.168.1.10', '192.168.1.11', '192.168.1.12']
for ip in ips:
  print('Adresse IP : ' + ip)
````

#### Parcours avec `enumerate`

- Une troisième option, qui combine les deux précédentes, est d'utiliser la fonction `enumerate` :

  - `enumerate(ma_liste)` génère une séquence de paires `(indice, element)` pour chaque élément de la liste

  - on peut donc à la fois accéder à l'indice et à l'élément dans la boucle `for`, en utilisant deux variables : `for index, element in enumerate(ma_liste):`

    - `index` reçoit l'indice de l'élément courant

    - `element` reçoit la valeur de l'élément courant

    - on peut nommer ces variables comme on le souhaite

````python
ips = ['192.168.1.10', '192.168.1.11', '192.168.1.12']
for index, ip in enumerate(ips):
    print('Indice ' + str(index) + ' : ' + str(ip))
````

- `enumerate()` est donc pratique quand on a besoin à la fois des indices et des éléments lors du parcours d'une liste

### Une valeur appartient-elle à une liste ? `in` et `not in`

- L'opérateur `in` a une autre fonction : il permet de vérifier si une valeur appartient à une liste

  - L'expression `une_valeur in ma_liste` renvoie `True` si `valeur` est un élément de `ma_liste`, et `False` sinon

  - on utilise typiquement cela dans des conditions (`if`, `while`, etc.)

- Il suffit d'utiliser l'opérateur `not in` pour vérifier qu'une valeur n'appartient pas à une liste

````python
ips = ['192.168.1.10', '192.168.1.11', '192.168.1.12']
ip_cible = '192.168.1.11'

if (ip_cible in ips):
  print('L\'IP ' + ip_cible + ' est dans la liste')

if ('192.168.1.10' not in ips):
  print ('L\'IP 192.168.1.25 n\'est pas dans la liste')
````

### Méthodes

- Une méthode est comme une fonction, mais _appelée sur_ une valeur, en utilisant la notation pointée

  - par exemple, on peut « demander » à une liste de trouver l'index d'un élément en utilisant la méthode `index()` : `ma_liste.index(valeur)`

- Le concept de _méthode_ n'est pas spécifique aux listes : chaque type de valeur (string, liste, etc.) possède ses propres méthodes

- Le type `list` possède de nombreuses méthodes pour manipuler les listes : recherche, ajout, suppression, tri...

#### Recherche d'un élément : `index()`

- La méthode `index(valeur)` permet de rechercher la position (indice) d'un élément dans une liste

  - elle renvoie l'indice de la _première occurrence_ de `valeur` dans la liste

  - si l'élément n'est pas trouvé, une exception `ValueError` est levée

  - on peut spécifier des indices de début et de fin optionnels pour restreindre la recherche à une partie de la liste : `ma_liste.index(valeur, debut, fin)`

````python
ips = ['192.168.1.10', '192.168.1.11', '192.168.1.11']
ip_cible = '192.168.1.11'
index_cible = ips.index(ip_cible)
print('L\'IP ' + ip_cible + ' se trouve à l\'indice ' + str(index_cible) + '.')
````

- Notez que l'index renvoyé est `1` (première occurrence), et pas `2`

#### Ajouter une valeur : `append()` et `insert()`

- La méthode `append(valeur)` permet d'ajouter un élément à la fin d'une liste : `ma_liste.append(nouvel_element)`

- La méthode `insert(indice, valeur)` permet d'insérer un élément à une position spécifique dans la liste : `ma_liste.insert(indice, nouvel_element)`

  - les éléments à partir de cet indice sont décalés d'une position pour faire de la place au nouvel élément

````python
ips = ['192.168.1.10', '192.168.1.11']
print(ips)
ips.append('192.168.1.12')  # Ajoute en fin de liste
print(ips)
ips.insert(1, '192.168.1.15')  # Insère à l'index 1
print(ips)
````

- Notez qu'on ne fait ici aucune affectation explicite, comme `ips = ips.append(...)`

  - la valeur renvoyée par `append()` et `insert()` est `None`
  
  - ces méthodes modifient la liste « _en place_ » sans créer une nouvelle liste

- Gardez à l'esprit que les méthodes appartiennent à un type bien précis : en règle générale, vous ne retrouverez pas les mêmes méthodes applicables sur d'autres types

  - une erreur de type `AttributeError` sera levée si vous essayez d'appeler une méthode qui n'existe pas pour un type donné

#### Supprimer une valeur : `remove()`

- On connaît déjà l'instruction `del` pour supprimer un élément d'une liste en connaissant son indice : `del ma_liste[indice]`

- La méthode `remove(valeur)` permet de supprimer un élément en connaissant sa valeur : `ma_liste.remove(element_a_supprimer)`

  - seule la _première occurrence_ trouvée sera supprimée

  - si l'élément n'est pas trouvé, une exception `ValueError` est levée

````python
ips = ['192.168.1.1', '192.168.1.2', '192.168.1.3', '192.168.1.2']
print(ips)
ips.remove('192.168.1.2') # Supprime la 1ère occurrence trouvée
print(ips)                # Notez que la 2ème occurrence est toujours présente
del ips[0]                # Supprime l'élément à l'index 0
print(ips)
````

#### Trier une liste : `sort()`

- La méthode `sort()` permet de trier les éléments d'une liste **en place**

  - par défaut, le tri est effectué dans l'**ordre croissant**

  - on peut spécifier l'argument optionnel `reverse = True` pour trier dans l'ordre décroissant : `ma_liste.sort(reverse = True)`

````python
liste = [3, 14, 3.14, 1, -14]
liste.sort()                # Tri de la liste (par défaut : ordre croissant)
print(liste)
liste.sort(reverse = True)  # Ordre décroissant doit être spécifié explicitement
print(liste)
````

- Quelques points à noter sur le tri avec `sort()` :

  - _tri en place_ : signifie que `sort()` modifie la liste elle-même, plutôt que d'en créer une nouvelle ; inutile d'essayer de récupérer la valeur de retour avec `sorted_list = liste.sort()`

  - pour les strings, l'ordre croissant est basé sur l'ordre lexicographique (ordre alphabétique), _sauf que_ les lettres majuscules sont triées avant les minuscules (donc `'Z' < 'a'`)

  - on ne peut pas trier une liste qui contient des éléments de types différents (ex : des strings et des entiers mélangés) ; cela lèvera une exception `TypeError`

````python
liste = ['Arnaud', 'affreux', 'Bébert', 'binouze', 'Casper', 'chichon']
liste.sort()  # Ordre ASCIIbetical => majuscules avant minuscules
print(liste)

liste = [1, 3, 2, 4, 'Pierre-Rémy']  # types mélangés => problème
liste.sort()
````

- On peut aussi fournir une fonction de clé personnalisée via l'argument `key` pour définir un ordre de tri spécifique
    
  - par exemple, pour forcer un tri de strings insensible à la casse, on peut fournir la fonction `str.lower` via l'argument `key`

  - cela aura pour effet de convertir chaque string en minuscules avant de les comparer pour le tri, mais sans modifier les valeurs originales dans la liste

````python
liste = ['Arnaud', 'affreux', 'Bébert', 'binouze', 'Casper', 'chichon']
liste.sort(key = str.lower)  # passage en lowercase juste pour le tri
print(liste)                 # les valeurs originales restent inchangées, mais triées
````

#### Renverser une liste : `reverse()`

- Si vous devez rapidement inverser l'ordre des éléments d'une liste, utilisez la méthode `reverse()`

  - la liste est modifiée _en place_

````python
liste = [10, 20, 30]
liste.reverse()       # Inversion en place des éléments
print(liste)
````

### Boule Magique - Version avec liste

- Reprenons l'exemple du script de la boule magique :

````python
import random

def magic_ball(answer_number):
  if answer_number == 1:
    return 'C\'est certain.'
  elif answer_number == 2:
    return 'C\'est sans aucun doute une possibilité'
  elif ... # code tronqué
  else:
    return 'L\'oracle a crashé. Veuillez réessayer plus tard.'

print('Posez une question de type oui/non : ')
input('>')
nb = random.randrange(1, 9)
answer = magic_ball(nb)
print(answer)
````

- Une liste permettrait ici de simplifier le code en stockant toutes les réponses possibles

   - au lieu d'avoir une longue série de conditions `if`/`elif` pour chaque réponse, on peut accéder directement à la réponse sélectionnée en utilisant un indice

````python
import random

reponses = [
  "C'est certain.",
  "C'est sans aucun doute une possibilité",
  "Oui",
  "Échec du moteur de décision. Concentrez-vous et recommencez.",
  "Ma réponse est non.",
  "Probablement pas.",
  "Cela reste incertain.",
  "ChatGPT ? Il dit encore des conneries."
]

def magic_ball(answer_number):
  return reponses[answer_number - 1]  # -1 car les indices commencent à 0

print('Posez une question de type oui/non : ')
input('>')
nb = random.randrange(1, 9)
answer = magic_ball(nb)
print(answer)
````

- Notez d'abord qu'une liste peut être écrite sur plusieurs lignes pour plus de clarté sans que cela pose aucun problème (on peut même indenter comme on veut)

- Cette méthode permet d'**améliorer la lisibilité** du code en évitant une longue série de conditions, et **facilite la maintenance** en permettant d'ajouter ou de modifier des réponses simplement en modifiant la liste

  - cette modification peut de plus être faite **dynamiquement** (pendant l'exécution) : par exemple, on pourrait imaginer charger les réponses depuis un fichier externe, ce qui permettrait de personnaliser les réponses sans modifier le code source du script

- La sélection aléatoire d'une réponse dans une liste est une opération courante, et elle a sa fonction dédiée : `random.choice()`

  - le script pourrait alors encore être simplifié en remplaçant complètement l'appel à la fonction par un appel à `random.choice()` (qui garantit de ne pas déborder au niveau des indices) :

````python
import random

reponses = [
  "C'est certain.",
  "C'est sans aucun doute une possibilité",
  "Oui",
  "Échec du moteur de décision. Concentrez-vous et recommencez.",
  "Ma réponse est non.",
  "Probablement pas.",
  "Cela reste incertain.",
  "ChatGPT ? Pas plus impressionné que ça."
]

print('Posez une question de type oui/non : ')
input('>')
answer = random.choice(reponses)
print(answer)
````

### Séquences

- Les listes ne sont pas le seul type de données en Python à représenter des collections ordonnées d'éléments : les strings, les objets renvoyés par `range`, les tuples (voir plus loin) en sont d'autres exemples

- Ces types de données sont appelés des **séquences** : des collections ordonnées d'éléments accessibles par des indices

  - comme ils ont tous ces caractéristiques en commun, ils partagent aussi plusieurs opérations : accès par indice, slices, parcours avec `for`, opérateurs `in` et `not in`...

````python
name = 'Kevin-Georges'
print(name[0])         # 'K'
print(name[-3])        # 'g'
print(name[0:5])       # 'Kevin'
print('Georg' in name) # True
print('kev' in name)   # False
print(' ' not in name) # True
for c in name:
  print('/// ' + c.upper() + ' \\\\\\')
````

### Types mutables et immuables

- Les listes et les strings diffèrent en revanche par un aspect fondamental : les listes sont **mutables**, tandis que les strings sont **immuables**

  - une valeur _mutable_ peut être modifiée _en place_ sans créer une nouvelle valeur ; par exemple, on peut modifier un élément d'une liste sans créer une nouvelle liste

  - une valeur _immuable_ ne peut pas être modifiée après sa création ; toute opération qui semble par exemple modifier une string crée en fait une nouvelle string, laissant l'originale inchangée

````python
liste = ['pomme', 'banane', 'cerise']
liste[1] = 'kiwi'
print(liste)
word = 'banane'
word[1] = 'i'  # Exception 'TypeError' : les strings sont immuables
````

- Pour « modifier » une string, on doit utiliser le _slicing_ et la concaténation et ainsi créer une nouvelle string

  - il est alors possible d'affecter cette nouvelle string à la même variable pour remplacer l'ancienne valeur ou utiliser une nouvelle variable pour conserver les deux versions

````python
name = 'Kevin-Georges'
name = name[:6] + 'Serge'   # ici on écrase l'ancienne valeur de 'name'
print(name)                 # 'Kevin-Serge'
````

#### Mutable vs. immuable : pour quoi faire ?

- La distinction entre types mutables et immuables peut sembler à première vue un détail technique

- Mais elle a des implications importantes en terme de **sécurité**, de **performance** et de **comportement dans les fonctions**

### Le type `tuple`

- Le type `tuple` est une autre structure de données qui ressemble aux listes

- Il y a deux principales différences avec les listes :

  - on utilise des parenthèses `( )` au lieu de crochets `[ ]` pour définir un tuple

  - les tuples sont **immuables** : une fois créés, leurs éléments ne peuvent pas être modifiés, ajoutés ou supprimés

- Utilisez donc les tuples lorsque vous voulez indiquer à Python (et aux autres lecteurs du script) que la séquence d'éléments ne doit pas être modifiée

- On accède aux éléments d'un tuple de la même manière que pour une liste : par indice, slices, etc.

````python
infos = ('PC-10', 'FF:EE:DD:CC:BB:AA', '192.168.1.10')
print(infos[0])        # 'PC-10'
print(infos[-1])       # '192.168.1.10'
infos[1] = '11:22:33:44:55:66'  # Exception 'TypeError' : tuple = immuable
````

- Si on n'a qu'une seule valeur dans le tuple, il faut ajouter une virgule après la valeur pour indiquer qu'il s'agit bien d'un tuple :

````python
small_tuple = (42,)  # tuple avec un seul élément
print(len(small_tuple))
````

### Références

- Dans la première partie, nous avions dit qu'une variable était comme une étiquette (nom de la variable) sur une boîte (contenant la valeur)

````python
carottes = 10
pommes = carottes
carottes = 20
print(carottes)  # 20
print(pommes)    # 10
````

- Quand on stocke la valeur `10` dans la variable `carottes` :

  - on crée la valeur `10` en mémoire
  
  - et on stocke l'**adresse** de cette valeur en mémoire dans la variable `carottes`
  
  - c'est ce qu'on appelle une **référence**

- Quand on écrit `pommes = carottes`, on ne crée pas une nouvelle valeur : on crée une nouvelle variable `pommes` qui **référence la même valeur** que `carottes` (on copie la référence)

  - donc, si on modifie `carottes` pour qu'elle référence une nouvelle valeur, `pommes` continue de référencer l'ancienne valeur

- Cela a des implications importantes pour les types mutables, comme les listes : si `carottes` et `pommes` référencent la même liste, une modification de la liste via `carottes` sera visible aussi via `pommes`, car les deux variables pointent vers la même liste en mémoire

````python
carottes = [10, 20, 30]
pommes = carottes   # on copie la référence, pas la liste
carottes[0] = 'coucou'
print(carottes)  # ['coucou', 20, 30]
print(pommes)    # ['coucou', 20, 30]
````

- Donc en Python, les variables ne _contiennent_ pas de valeurs : elles contiennent des _références_ à des valeurs
- L'opérateur d'affectation (`=`) copie juste la référence ; il ne copie jamais de valeurs

#### Références et arguments de fonction

- Quand une fonction est appelée, Python **copie dans les variables paramètres les références** aux valeurs passées en arguments

  - quand un argument est de type mutable (ex : une liste), ça veut dire que la fonction peut modifier la valeur référencée, et **ces modifications seront visibles en dehors de la fonction**

````python
def append_element(some_list, some_element):
  some_list.append(some_element)

carottes = [10, 20, 30]
append_element(carottes, 40)
print(carottes)  # [10, 20, 30, 40]
````

- Notez qu'on n'a pas besoin d'une écriture comme `carottes = append_element(carottes, 40)` dans l'exemple précédent :

  - la fonction `append_element` modifie la liste référencée par `carottes` _en place_

  - la variable `carottes` et le paramètre `some_list` sont deux références distinctes, mais elles pointent vers la même liste en mémoire

#### `copy()` et `deepcopy()`

- La plupart du temps, ce comportement par défaut est ce qu'on veut : il permet d'écrire des fonctions qui modifient des structures de données complexes sans avoir à les copier entièrement

- Quand on souhaite vraiment créer une copie indépendante d'une structure mutable (comme une liste), Python fournit le module `copy` avec deux fonctions :

  - La méthode `copy()` : crée une copie « superficielle » (_shallow copy_) de la liste (on copie les références des éléments de la liste dans la nouvelle liste)
  - La fonction `deepcopy()` : crée une copie « profonde » (_deep copy_) : toutes les références imbriquées sont également copiées, créant une structure entièrement indépendante

````python
import copy
carottes = [10, 20, 30]
pommes = copy.copy(carottes)  # shallow copy (superficielle)
pommes[0] = 'coucou'
print(carottes)  # [10, 20, 30] => reste inchangée
print(pommes)    # ['coucou', 20, 30]

# Si la liste copiée contient des listes internes, elles ne sont pas dupliquées
# Une modif d'une liste interne affecte les deux listes
carottes = [[1, 2, 3], [4, 5, 6]]
pommes = copy.copy(carottes)  # shallow copy (superficielle)
pommes[0][0] = 'coucou'
print(carottes)  # [['coucou', 2, 3], [4, 5, 6]] => liste interne modifiée
print(pommes)    # [['coucou', 2, 3], [4, 5, 6]]

# Si on veut une copie complète, on utilise deepcopy
carottes = [[1, 2, 3], [4, 5, 6]]
pommes = copy.deepcopy(carottes)  # deep copy (complète)
pommes[0][0] = 'coucou'
print(carottes)  # [[1, 2, 3], [4, 5, 6]] => reste inchangée
print(pommes)    # [['coucou', 2, 3], [4, 5, 6]]
````

### Synthèse

- 1 liste = permet de travailler, _à partir d'une seule variable_, sur un ensemble d'éléments
- Les éléments sont ordonnés, accessibles par indice, et peuvent être de types différents (en général non)
- On peut appeler des **méthodes** sur une liste pour effectuer des opérations courantes : recherche, ajout, suppression, tri...
- Liste = mutable, tuple = immuable, string = immuable
- Une variable contient une référence à une valeur, pas la valeur elle-même => une variable ne peut pas « contenir » toute une liste : elle _référence_ la liste
- Donc quand on affecte une variable à une autre, on copie la référence, pas la valeur : les deux variables pointent vers la même valeur en mémoire
- Cela a des implications importantes quand on passe une liste (mutable) à une fonction : la fonction peut modifier la liste, et ces modifications seront visibles en dehors de la fonction
- On peut effectuer des copies de listes avec `copy()` et `deepcopy()`
